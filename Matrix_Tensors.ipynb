1. Melakukan berbagai operasi yg diberikan dengan numpy

```python
import numpy as np
```

```python
# buat matrix X berukuran 2x3 menggunakan numpy
x = np.array([[1, 5, 11], [55, 15, 55]])
x
```

```python
# tampilkan ukuran dari matrix X
x.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x[:, 0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x[1, :]
```

2. Melakukan Operasi yg telah dilakukan menggunakan pytorch dan tensorflow

```python
import torch
import tensorflow as tf
```

```python
# buat matrix x_t berukuran 2x3 menggunakan pytorch
x_t = torch.tensor([[1, 5, 11], [55, 15, 55]])
x_t
```

```python
# tampilkan ukuran dari matrix x_t
x_t.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X_t
x_t[:, 0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X_t
x_t[1, :]
```

```python
# buat matrix x_t berukuran 2x3 menggunakan tensorflow
x_tf = tf.Variable([[1, 5, 11], [55, 15, 55]])
x_tf
```

```python
# tampilkan ukuran dari matrix x_tf
x_tf.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X_tf
x_tf[:, 0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X_tf
x_tf[1, :]
```
